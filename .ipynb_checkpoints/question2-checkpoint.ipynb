- For both survivals and casualties plot their age and ticket fare on one scatter plot. Show different marker for each split as shown in Figure 1.

The dataset contains the following features:

```python
# You solution goes here
# you are encouraged to use multiple cells in the notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# getting the dataset
data = pd.read_csv('train.csv')

# splitting the dataset into survived and not survived
survived = data[data['Survived'] == 1]
not_survived = data[data['Survived'] == 0]

# Correcting the scatter plot with Fare on the X-axis and Age on the Y-axis
plt.figure().set_figwidth(6)

plt.scatter(survived['Fare'], survived['Age'], c='blue', label='Survived', marker='v')
plt.scatter(not_survived['Fare'], not_survived['Age'], c='orange', label='Casualties', marker='o')

plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend()
plt.show()

# Correcting the histograms for the count of passengers from each embarkation port
# with separated bars for clarity and using full names of embarkation ports

# Histogram for survivors based on Embarkation port
colors = ['blue', 'orange', 'green']
plt.figure(figsize=(10, 6))
survived_counts = survived['Embarked'].value_counts().reindex(['C', 'Q', 'S']).fillna(0)
plt.bar(survived_counts.index, survived_counts.values, color=colors, label='Survived', width=0.1, align='center')
plt.xlabel('Port')
plt.ylabel('Surived')
plt.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
plt.show()

# Histogram for casualties based on Embarkation port
plt.figure(figsize=(10, 6))
not_survived_counts = not_survived['Embarked'].value_counts().reindex(['C', 'Q', 'S']).fillna(0)
plt.bar(not_survived_counts.index, not_survived_counts.values, color=colors, label='Casualties', width=0.1, align='center')
plt.xlabel('Port')
plt.ylabel('Casualties')
plt.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
plt.show()
```
